```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import BernoulliNB
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
dataset = pd.read_csv('/home/bk/Desktop/mlai/Notes/Day 24/train.csv')
dataset.head()
```

```python
dataset.info()
```

```python
dataset.isna().sum()
```

```python
dataset.drop(['Cabin','PassengerId','Name','Ticket'],axis = 1, inplace = True)
```

```python
dataset.columns
```

```python
dataset.head()
```

```python
le = preprocessing.LabelEncoder()
dataset['Sex'] = le.fit_transform(dataset['Sex'])
dataset['Embarked'] = le.fit_transform(dataset['Embarked'])
dataset.head()
```

```python
dependent_cols = ['Pclass' , 'Sex', 'SibSp' , 'Parch', 'Embarked']
for col in dependent_cols:
    X = dataset.drop(col,axis=1)
    y = dataset[col]
    X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3,random_state=0)
    clf=BernoulliNB()
    y_pred=clf.fit(X_train,y_train).predict(X_test)
    print('===============================================================')
    print('Dependent Variable : '+col+'\n\n')
    print("Accuracy score is:",accuracy_score(y_test,y_pred,normalize=True))
    print('Confusion Matrix:')
    print(confusion_matrix(y_test, y_pred))
    print('===============================================================')
```
